## Restrcuture

```python
we = pd.read_csv('Data/weather.csv')
```

```python
we.head()
```

```python
wem = pd.melt(we, id_vars=['year', 'month', 'element'], var_name='day', value_name='temp')
```

```python
wem.shape
```

```python
we_tidy = wem.pivot_table(index =['year', 'month', 'day'], 
                columns='element', 
                values='temp',dropna=False).reset_index()
```

```python
we_tidy.to_csv('Data/Weather_tidy.csv') #path should exist
```

## Custom

```python
def mean_func(x):
    return sum(x) / len(x)
```

```python
we_tidy[['min', 'max']].apply(mean_func, axis=1)
```

## Merge


```python
users = pd.DataFrame({'id': [1,2,3], 'name':['A','B','C']})
users
```

```python
msgs = pd.DataFrame({'userid': [1,1,2], 'msg':['hello', 'hi', 'bye']})
msgs
```

```python
users.merge(msgs, left_on='id', right_on='userid')
```
