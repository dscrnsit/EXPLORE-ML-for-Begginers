## Styling

```python
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]
plt.plot(x,y, 'b^')# b is blue can use any colour o means dot |||ly many more.
plt.show()
```

```python
import numpy as np
x = np.arange(1, 6)
y = x**3
x2 = np.arange(1,6)
y2 = x2**2
plt.title('Cubes')# title for plot
plt.xlabel('X-axis')# x labels for plot
plt.ylabel('Cubes')
plt.xlim(1,6) #sets range for x axis
plt.ylim(0,130)#sets range for y axis
plt.plot(x,y,'r', label='Cubes')
plt.plot(x2,y2, label='Squares')
plt.legend()#shows labels
plt.grid()#shows grid
plt.text(3, 20, 'This') #shows text on plot
plt.show()
```

## Graded Assignment

```python
import pandas as pd
```

```python
df = pd.read_csv('Data\gapminder.csv')
df.head()
india = df[df.country=='India']
```

```python
year = list(india.year)
lifeExp = list(india.lifeExp)
pop = np.array(india.population)
pop = pop/10000000
gdp = np.array(india.gdpPerCap)
gdp = gdp/10
```

```python
plt.title('India Stats')
plt.xlabel('Years')
plt.ylabel('Stats')
plt.xlim(1950, 2010)
plt.ylim(40,250)
plt.plot(year, lifeExp,'b', label='LifeExpectancy')
plt.plot(year, pop,'k', label='Population(inM)')
plt.plot(year, gdp,'g',label='GDP(*10)')
plt.legend()
plt.grid()
plt.savefig('India_Stat.png', bbox_inches='tight')
```
